```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris
```

```python
df =sns.load_dataset("iris")
```

```python
# Independent feature and dependent feacture
x = df.iloc[:,:-1]
y = iris.target
```

```python
x,y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
Tree_model = DecisionTreeClassifier()
```

```python
Tree_model.fit(x_train,y_train)
```

```python
# we use gini
from sklearn import tree
plt.figure(figsize=(15,10))

tree.plot_tree(Tree_model,filled = True)
plt.show()
```

Above Decision tree at x[2] we have values(34 and 4),(1 and 30) which means huge difference so we can use here post pruning and stop decision tree using "max_depth" highper parameter.

```python
# PostPruning
Tree_model = DecisionTreeClassifier(max_depth=2)
```

```python
Tree_model.fit(x_train,y_train)
```

```python
# we use gini
from sklearn import tree
plt.figure(figsize=(15,10))

tree.plot_tree(Tree_model,filled = True)
plt.show()
```

```python
# Prediction
y_pred = Tree_model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
score = accuracy_score(y_pred,y_test)
score
```

```python
print(classification_report(y_pred,y_test))
```
