3. Split the Data: Split the dataset into training and testing sets using an 80-20 split.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.drop(["Disease"],axis="columns"),df["Disease"],test_size = 0.2)
```

```python
print(len(x_train))
print(len(x_test))
print(len(y_train))
print(len(y_test))
```

4. Train a Decision Tree Model: Train a Decision Tree classifier on the training data to predict whether a patient has heart disease.

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

5. Evaluate the Model: Evaluate the performance of the model using appropriate metrics such as accuracy, precision, recall, and the confusion matrix.

```python
y_predi = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
accuracy = accuracy_score(y_test, y_predi)
print(f"Accuracy: {accuracy:.2f}")
precision = precision_score(y_test, y_predi)
print(f"Precision: {precision:.2f}")
recall = recall_score(y_test, y_predi)
print(f"Recall: {recall:.2f}")
cm = confusion_matrix(y_test, y_predi)
print("Confusion Matrix:")
print(cm)
```

```python
plt.figure(figsize=(8,5))
sb.heatmap(cm, annot=True, fmt=".0f", cbar_kws={'format': '%.0f'})
plt.xlabel("Predicted")
plt.ylabel("Truth")
plt.show()
```

6. Visualize the Decision Tree: Visualize the trained Decision Tree to understand the decision-making process of the model.

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(15, 10))
plot_tree(model, filled=True, feature_names=x_test.columns, class_names=True)
plt.show()
```

7. Interpret the Results: Provide an interpretation of the results. Discuss the most important features contributing to the prediction of heart disease and any potential overfitting or underfitting issues with the model.

```python
feature_importance = model.feature_importances_
```

```python
features = x_test.columns
importance_df = pd.DataFrame({'Feature': features, 'Importance': feature_importance})
importance_df = importance_df.sort_values(by='Importance', ascending=False)
```
